## Multiple ways to call the dunder(double underscore) (also known as magic methods) method \_\_add\_\_

```python
str.__add__("Data", " Analyst")
```

```python
"Data".__add__(" Analyst")
```

```python
# + operator uses __add__ method under the hood
"Data" + " Analyst"
```

## Other string dunder methods

```python
#__mul__
"Data Analyst" * 10
```

```python
help(str.__len__)
```

```python
#__len__
len("Data Analyst")
```

# Problems

## Convert to Uppercase (1.4.1) - Problem

```python
job_title = 'data scientist'

job_title_upper = job_title.upper()

print(job_title)
```

## Replace Substring (1.4.2) - Problem

```python
job_title = 'Data Analyst'

job_title_engineer = job_title.replace('Analyst', "Engineer")

print(job_title_engineer)
```

## Concatenate Strings (1.4.3) - Problem

```python
seniority_level = 'Senior'
job_title = 'Data Analyst'

senior_data_analyst_title = seniority_level + ' ' + job_title
print(senior_data_analyst_title)
```

## Check String Length (1.4.4) - Problem

```python
job_title = 'Data Scientist'
print(len(job_title))
```

## Convert to Lowercase (1.4.5) - Problem

```python
job_title = 'DATA ENGINEER'

job_title_lower = job_title.lower()

print(job_title_lower)
```
