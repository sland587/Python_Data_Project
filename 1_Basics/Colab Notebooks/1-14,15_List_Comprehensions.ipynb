# List Comprehensions

````python
# List Comprehension example
numbers = [x for x in range(100)]

print(numbers)
````

````python
# Manual way of doing it
numbers = []

for number in range(100):
  numbers.append(number)

print(numbers)
````

## List Comprehension breakdown

Syntax for List Comprehension:
[Expression for item in iterable]


1.   Define List
```
numbers = []
```

2.   Define variable to put into list which can be acted upon

```
numbers = [x]
```

3.   Add loop referencing variable defined above

```
numbers = [x for x in range(100)]
```

4.   How to Read

For each iteration of the loop, assign the item to x and then add x to the list.



## Acting on Defined Variable example

````python
# List Comprehension example 2
# Act on variable before storing in list
numbers = [x*2 for x in range(100)]

print(numbers)
````

## Type conversion example

````python
# List Comprehension example 3
# Convertion of type example
numbers = [float(x*2) for x in range(100)]

print(numbers)
````

# Exercise

Filter what jobs to apply to based on a list of skills

[Exercise - Basics](https://github.com/lukebarousse/Python_Data_Analytics_Course/blob/main/1_Basics/15_Exercise_Basics.ipynb)

````python
# Define data science job roles and required skills
job_roles = [
    {'role': 'Data Analyst', 'skills': ['Python', 'SQL', 'Excel']},
    {'role': 'Data Scientist', 'skills': ['Python', 'R', 'Machine Learning', 'Deep Learning']},
    {'role': 'Machine Learning Engineer', 'skills': ['Python', 'TensorFlow', 'PyTorch', 'Scikit-Learn']},
    {'role': 'Data Engineer', 'skills': ['Python', 'Apache Spark', 'Hadoop', 'SQL']},
    {'role': 'Business Intelligence Analyst', 'skills': ['Python', 'SQL', 'Tableau', 'Power BI', 'Excel']},
    {'role': 'Quantitative Analyst', 'skills': ['R', 'Python', 'MATLAB', 'Statistics']},
    {'role': 'Operations Analyst', 'skills': ['Python', 'SQL', 'Data Visualization', 'Process Improvement']},
    {'role': 'Database Administrator', 'skills': ['SQL', 'Oracle', 'MySQL', 'Database Management']},
    {'role': 'AI Engineer', 'skills': ['Python', 'TensorFlow', 'PyTorch', 'Computer Vision']},
    {'role': 'Statistician', 'skills': ['R', 'SAS', 'Python', 'Statistical Modeling']}
]

# My skills
my_skills = ['Python', 'SQL', 'Excel']
````

## Pseudocode


1.   Loop through the `job_roles`
1.  Verify if all `my_skills` are inside each `job_roles['skills']`
1.   if `True` (or qualifed), add to qualified list of jobs; Note: Can have more skills required as long as all my skills are listed.






````python
# 1. Loop through the job_roles
for job in job_roles:
  print(job)
````

````python
qualified_roles = []

for job in job_roles:
  qualified = True

  for skill in my_skills:
    # 2. Verify if all my_skills are inside each job_roles['skills']
    if skill not in job['skills']:
      qualified = False
      break

  # 3. If `True` (or qualifed), add to qualified list of jobs
  if qualified:
    qualified_roles.append(job['role'])


print(qualified_roles)
````

````python
# Rewrite with `all` method which uses comprehension inside of `all method`

# Determine which jobs you're qualified for
qualified_roles = []

for job in job_roles:
  # Check if all required skills are in my_skills
    if all(skill in job['skills'] for skill in my_skills):
      # Add the job role to the qualified_roles list
      qualified_roles.append(job['role'])

print(qualified_roles)
````

````python
# Full List comprehension form
qualified_roles = [job['role'] for job in job_roles if all(skill in job['skills'] for skill in my_skills)]

print(qualified_roles)
````
