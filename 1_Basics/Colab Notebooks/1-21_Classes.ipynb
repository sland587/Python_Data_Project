# Classes

```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate=0.1, symbol="$") -> None:
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol
    self.total_salary = self.base_salary * (1 + self.bonus_rate)
    self.bonus = self.total_salary - self.base_salary

  def __repr__(self) -> str:
    return f'{self.symbol}{self.base_salary:,.0f}'

  def show_salary(self):
      return f'{self.symbol}{self.total_salary:,.0f}'

  def show_bonus(self):
      return f'{self.symbol}{self.bonus:,.0f}'

```

```python
salary = BaseSalary(100000)

print(salary.base_salary, salary.bonus_rate, salary.symbol)
```

```python
salary
```

```python
salary.show_salary()
```

```python
salary.show_bonus()
```

# Problems

```python
# Class which is solution for all problems
class JobPosting:
  def __init__(self, title, company, location, salary = 100_000) -> None:
     self.title = title
     self.company = company
     self.location = location
     self.salary = salary

  def __repr__(self) -> str:
     return f'Title: {self.title}, Company: {self.company}, Location: {self.location}, Salary: {self.salary}'

  def to_dict(self):
    return {
        'title':self.title,
        'company':self.company,
        'location':self.location,
        'salary':self.salary
        }

  def compare_salary(self, job_posting):
    highest_job_title = job_posting.title

    if(self.salary > job_posting.salary):
      highest_job_title = self.title

    return highest_job_title
```

## Create JobPosting Class (1.21.1) - Problem

```python
job = JobPosting('Data Scientist', 'Tech Innovations', 'New York')

print(job)
```

## Convert to Dictionary (1.21.2) - Problem

```python
job = JobPosting('Data Scientist', 'Tech Innovations', 'New York', 120000)

job.to_dict()

```

## Compare Job Salaries (1.21.3) - Problem

```python
job1 = JobPosting('Data Scientist', 'Tech Innovations', 'New York', 120000)
job2 = JobPosting('Data Analyst', 'Data Driven Co', 'San Francisco', 100000)

job1.compare_salary(job2)
```
