## Slicing Array (1.22.2) - Problem

```python
import numpy as np
postings_list = [10, 15, 7, 20, 25, 30, 5]

postings = np.array(postings_list)

first_3_postings = postings[:3]
print(first_3_postings)
```

## Highest and Lowest Salary (1.22.3) - Problem

```python
import numpy as np

salaries_list = [70000, 85000, 60000, 95000, 80000]

salaries = np.array(salaries_list)

minimum = np.min(salaries)
maximum = np.max(salaries)

print(minimum)
print(maximum)
```
