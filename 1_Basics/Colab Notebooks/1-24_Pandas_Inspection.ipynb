```python
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')

# Put Data into a data frame(df)
df = dataset['train'].to_pandas()

```

## DataFrame

### `head()`

```python
# Default prints first 5 rows, if row count int not provided
df.head()
```

```python
# Print first n rows where n = 10 rows in this case
df.head(10)
```

### `tail()`

```python
# Prints last 5 by default
df.tail()
```

```python
# print last n rows where n is 10 in this case
df.tail(10)
```

### specify column

```python
# dataFrame['columnName']
df['job_title_short']
```

```python
# Get first 10 rows of a certain column
df['job_title_short'].head(10)
```

```python
# Multiple Columns(must provide list which requires two sets of []; Otherwise, interpreted as a tuple and errors out)

df[['job_title_short', 'job_location']]
```

```python
# head and tails still works with multiple columns
df[['job_title_short', 'job_location']].tail(10)
```

### iloc - Location Indexer

```python
# Get all columns for row 90
df.iloc[90]
```

```python
## slicing - get rows 90 thru 99
df.iloc[90:100]
```

```python
# slicing - get rows 90 thru 99 and the first(0) and second column(1) only
df.iloc[90:100, 0:2]
```

```python
# Combining Multiple Columns with row slicing(better approach)

df[['job_title_short', 'job_title']].iloc[90:100]
```

### `info()`

```python
#gets metadata about the data frame
df.info()
```

### `describe()`

```python
# Provides statistical values for only the numerical columns
df.describe()
```

## DataFrame Column

### `unique()`

```python
# Returns an array of unique items in the data frame column
df.job_title_short.unique()
```

### column filters

```python
# Find rows that have a job title short = Data Analyst
df.job_title_short == 'Data Analyst'
```

```python
# Use the row filter to grab only rows that match the criteria
df[df.job_title_short == 'Data Analyst']
```

```python
# Filter on multiple columns( use '&' and '|' bitwise operator instead of 'and' and 'or' logical operator)
df[(df.job_title_short == 'Data Analyst') & (df.salary_year_avg.notna())]
```
