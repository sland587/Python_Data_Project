## Arithmetic Operators

Arithmetic operators are used with objects (mainly numeric ones) to perform math operations:

| Operator 	| Name           	| Example 	|
|----------	|----------------	|---------	|
| +        	| Addition       	| x + y   	|
| -        	| Subtraction    	| x - y   	|
| *        	| Multiplication 	| x * y   	|
| /        	| Division       	| x / y   	|
| %        	| Modulus        	| x % y   	|
| **       	| Exponentiation 	| x ** y  	|
| //       	| Floor division 	| x // y  	|

```python
2 + 3.0
```

```python
"What's " + "Up" + "!"
```

```python
"What's Up? " * 10
```

```python
3.0 - 2
```

```python
"What's Up" - "up"
```

```python
four = 4

four.__sub__(2)
```

```python
4 - 2
```

```python
5 / 2
```

```python
# Floor Division ( Division with whole number with float result or remainders provided)
5 // 2
```

```python
# Modulus to get remainder
5 % 2
```

## Order of Operations

1. **Parentheses**: Anything in parentheses is executed first
2. **Exponents**: Operations involving exponents (including roots, indicated as fractional exponents)  
3. **Multiplication and Division**: Processed from left to right
4. **Addition and Subtraction**: Also processed from left to right

```python
min_salary = 20_000 # _ is ignored by the interpreter
max_salary = 80_000

avg_salary = (min_salary + max_salary) / 2

print(avg_salary)
```

## Assignment Operators

Assignment operators are used to assign values to variables.

| Operator | Name                | Example |
|----------|---------------------|---------|
| =        | Assign              | x = y   |
| +=       | Add and assign      | x += y  |
| -=       | Subtract and assign | x -= y  |
| *=       | Multiply and assign | x *= y  |
| /=       | Divide and assign   | x /= y  |
| %=       | Modulus and assign  | x %= y  |
| **=      | Exponent and assign | x **= y |
| //=      | Floor divide and assign | x //= y  |

```python
applicant_1 = "Kelly"
applicant_2 = "Luke"

# Reverse applicant where applicant_1 is applicant_2 and applicant_2 is applicant_1

applicant_1, applicant_2 = applicant_2, applicant_1

print(applicant_1, applicant_2)
```

```python
x = 1

# x = x + 1
x += 1

print(x)
```

```python
x = 1

# x = x * 2
x *= 2

print(x)
```

## Comparison Operators

Comparison operators compare two values and determine the relation between them.

| Operator | Name                  | Example |
|----------|-----------------------|---------|
| ==       | Equal to              | x == y  |
| !=       | Not equal to          | x != y  |
| >        | Greater than          | x > y   |
| <        | Less than             | x < y   |
| >=       | Greater than or equal | x >= y  |
| <=       | Less than or equal    | x <= y  |


```python
salary_kelly = 60_000
salary_luke = 40_000

print(salary_kelly == salary_luke)
```

```python
print(1 == 1)
```

```python
1 = 1
```
