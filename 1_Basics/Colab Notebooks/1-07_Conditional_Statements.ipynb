# Conditional Statements

## Overview

### Notes

`if`, `elif`, and `else` are conditional statements that control the flow of execution based on whether a condition is `True` or `False`.

| Part       | Purpose                           | Condition Evaluated             | Action if True                 |
|------------|-----------------------------------|---------------------------------|--------------------------------|
| `if`       | First condition check             | First condition to be evaluated | Executes block if condition is true |
| `elif`     | Subsequent condition check        | Next condition, if `if` is false| Executes block if this condition is true |
| `else`     | Fallback if no conditions are met | None | Executes block if all above conditions are false |


```python
if True:
  print("What's up Data Nerds")

if False:
  print("What's not up Data Nerds")
```

```python
# Skill required for job postings
job_skill = 'SQL'

# Applicant Skill
applicant_skill = 'SQL'

if(applicant_skill == job_skill):
  print("Skills Match!")
```

```python
job_skill = 'SQL'
applicant_skill = 'Python'
years_experience = 3

if(applicant_skill == job_skill):
  print("Skills Match!")
elif years_experience >= 5:
  print('Enough Experience; No SQL')
elif applicant_skill == 'Python':
  print('This applicant should know SQL')
else:
  print("No skill or Experience")
```

##  Determine Job Title (1.7.3) - Problem

```python
job_experience = 4
sql_experience = 2

if job_experience < 2: # Less than 2 years of job experience
  if sql_experience == 0:
    job_title = 'Associate Analyst'
  else:
    job_title = 'Junior Analyst'
elif job_experience <= 5: # 2 to 5 years of job experience
  if sql_experience == 0:
    job_title = 'Associate Analyst'
  elif sql_experience == 1:
    job_title = 'Intermediate Analyst'
  else:
    job_title = 'Senior Analyst'
else: # More than 5 years of job experience
  if sql_experience <= 2:
    job_title = 'Data Analyst'
  else:
    job_title = 'Lead Analyst'

print(job_title)
```
