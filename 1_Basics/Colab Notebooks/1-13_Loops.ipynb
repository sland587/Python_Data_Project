# Loops

## `for` loops

```python
numbers = [0, 1, 2, 3, 4, 5]

for number in numbers:
  print(number)
```

```python
characters = 'PYTHON'

for character in characters:
  print(character)
```

```python
years_experience = {
    'Luke': 3,
    'Kelly': 6,
    'Ken': 4,
    'Alex': 3
}

# keys only
for person in years_experience:
  print(person)

print()

# values only
for value in years_experience.values():
  print(value)

print()

# key/value pair
for key, value in years_experience.items():
  print(key, value)

print()

# Candidates with > 5 years experience
print("Candidates with > 5 years experience: ")

# key/value pair
for key, value in years_experience.items():
  if( value > 5 ):
    print(key, 'has', value, 'years of experience.')
```

```python
for i in range(1,11):
  print(i)
```

## `while` loops

```python
i = 0

while i < 10:
  print(i)

  i += 1
```

```python
years_experience = {
    'Luke': 3,
    'Kelly': 6,
    'Ken': 4,
    'Alex': 3,
    'Mary': 7
}

years = 0
index = 0

years_list = list(years_experience.items())

print(years_list)

while years < 5:
  key, years = years_list[index]
  if (years >= 5):
      print(key, 'has', years, 'years of experience.')
  index += 1

```

## Print Job Titles (1.13.1) - Problem

```python
job_titles = ['Data Scientist', 'Data Analyst', 'Machine Learning Engineer']

for job_title in job_titles:
  print(job_title)
```

## Total Job Postings (1.13.2) - Problem

```python
job_postings = [5, 10, 15, 20, 25]

index = 0
number_of_job_postings = 0

while index < len(job_postings):
  number_of_job_postings += job_postings[index]
  index += 1

print('Total Number of Jobs:', number_of_job_postings)
```

## Print Job Titles with Index (1.13.3) - Problem

```python
job_titles = ['Data Scientist', 'Data Analyst', 'Machine Learning Engineer']

for index, title in enumerate(job_titles):
  print(f'{index}: {title}')
```

## Print Job Titles Containing 'Data' (1.13.5) - Problem

```python
positions_skills = {
    'Data Scientist': ['Python', 'R'],
    'Data Analyst': ['SQL', 'Excel'],
    'Machine Learning Engineer': ['Python', 'TensorFlow']
}

for position, skills in positions_skills.items():
  for skill in skills:
    print('The role of', position, 'requires proficiency in', skill)
  print()
```
