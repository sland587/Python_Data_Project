# Pandas: Index Management

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## DataFrame.index review

```python
df.head(5)
```

```python

df.index
```

```python
# Give the index a column name
df.index.name = 'job_index'

df.head(5)
```

```python
df.index.dtype
```

## [DataFrame.reset_index()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reset_index.html)

```python
df_usa = df[df['job_country'] == 'United States']

# indexes are off because subset of original data
df_usa.head()
```

```python
# set inplace to true so we modfiy the df_usa dataframe instead of creating a new one.abs
df_usa.reset_index(inplace=True)

df_usa
```

## [DataFrame.set_index()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.set_index.html)

```python
# set the index instead to the the originally named job_index
df_usa.set_index(keys='job_index', inplace=True)

df_usa
```

## [DataFrame.sort_index](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_index.html)

```python
# better use case for this is sorting a pivot_table

median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])
median_pivot
```

```python
# For reference, how to sort by value using first level and second level column labels
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```

```python
# Finally demonstrate sort by index
median_pivot.sort_index(inplace=True)
median_pivot
```

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Set Index and Display Rows (2.5.1) - Problem

```python
df_problem1 = df.copy()
df_problem1.set_index(keys='job_posted_date', inplace=True)
df_problem1.head()
```

## Reset Index and Drop Existing (2.5.2) - Problem

```python
# Set the starting point as the end point from the first problem
df_problem2 = df_problem1.copy()

df_problem2.reset_index(drop=True, inplace=True)
df_problem2.head()
```

## Sort Index (2.5.3) - Problem

```python
median_salary_pivot = df.pivot_table(values='salary_year_avg', index='company_name', aggfunc='median')

median_salary_pivot.sort_index(inplace=True, ascending=False)
median_salary_pivot.head()
```

## Pivot Table Reset & Set Index (2.5.4) - Problem

```python
median_salary_by_job_title_pivot = df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
median_salary_by_job_title_pivot.reset_index(inplace=True)
median_salary_by_job_title_pivot.set_index(keys='salary_year_avg', inplace=True)
median_salary_by_job_title_pivot.head()
```
