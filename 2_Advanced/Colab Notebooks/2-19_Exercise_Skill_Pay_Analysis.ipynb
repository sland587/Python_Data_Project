# Exercise: Skill Pay Analysis

```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Only get data analyst jobs in the US
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')] .copy()

# Drop NaN values from the 'salary_year_avg' column for accurate visualization
df_DA_US = df_DA_US.dropna(subset=['salary_year_avg'])

df_DA_US
```

```python
# Explode out the job skills column
df_DA_US = df_DA_US.explode('job_skills')

df_DA_US['job_skills']
```

```python
# Get count and median yearly salary by job skill

df_DA_US_group = df_DA_US.groupby(['job_skills'])['salary_year_avg'].agg(['count', 'median'])
df_DA_US_group
```

```python
# Get Top Paying Skills
df_DA_top_pay = df_DA_US_group.sort_values(by='median', ascending=False).head(10)

# Get highest count (demand) of skills and highest median salary by skill
df_DA_skills = df_DA_US_group.sort_values(by='count', ascending=False).head(10)
df_DA_skills = df_DA_skills.sort_values(by='median', ascending=False)
```

```python
# Plot both the top paying skills and the skills with top job count
rows_count = 2
columns_count = 1
fig, ax = plt.subplots(rows_count, columns_count)

# Plot top paying skills
df_DA_top_pay[::-1].plot(kind='barh', y='median', ax=ax[0], legend=False)
ax[0].set_title('Top 10 Highest Paid Skills for Data Analysts')
ax[0].set_ylabel('')
ax[0].set_xlabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

# Plot highest count (demand) of skills and highest median salary by skill
df_DA_skills[::-1].plot(kind='barh', y='median', ax=ax[1], legend=False)
ax[1].set_xlim(ax[0].get_xlim())
ax[1].set_title('Top 10 Most In-Demand Skills for Data Analysts')
ax[1].set_ylabel('')
ax[1].set_xlabel('Median Salary ($USD)')
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

fig.tight_layout()
plt.show()
```
