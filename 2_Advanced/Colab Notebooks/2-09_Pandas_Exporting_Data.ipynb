````python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
````

## Example Export

### Setup data for export

````python
#Create column to identify data for each month; Use 3 letter abbreviation of the month
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')

# Get unique list of months
months = df['job_posted_month'].unique()

# Use Dictionary Comprehension to seperate the data by each month
# Syntax
# {key: value for value in value_list}
dict_months = {month: df[df['job_posted_month'] == month] for month in months}

# Combine the first quarter's worth of data
df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)
````
