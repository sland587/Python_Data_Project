## [df.copy()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.copy.html)

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()

df_original
```

```python
df_altered = df_original.copy()
```

```python
print('ID of df_original:               ', id(df_original))
print('ID of df_altered:                ', id(df_altered))
print('Are the two dataframes the same? ', id(df_original) == id(df_altered))
```

```python
median_salary = df_altered['salary_year_avg'].median()

df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)

print('Altered:')
print(df_altered.loc[:5, 'salary_year_avg'])

print('Original:')
print(df_original.loc[:5, 'salary_year_avg'])
```

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Drop Missing Values (2.3.1) - Problem

```python
df_copy = df.copy()

df_copy = df_copy.dropna(subset='salary_year_avg')
df_copy.head()
```

## Random Sample of DataFrame (2.3.2) - Problem

```python
df_sample = df.sample(frac=.1)

df_sample.head()
```

## Fill Missing Salaries (2.3.3) - Problem

```python
df_updated = df.copy()

median_salary = df_updated['salary_year_avg'].median()

df_updated['salary_year_avg'] = df_updated['salary_year_avg'].fillna(median_salary)

df_updated.head()
```
